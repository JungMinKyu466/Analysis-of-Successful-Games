```python
df.columns = df.iloc[0]     # 0번 행을 컬럼으로 설정
df = df[1:]                 # 0번 행 제거
df = df.reset_index(drop=True)  # 인덱스 재설정
df.head()
```

```python

df.columns= ['Name', 'Release date', 'Estimated owners', 'Peak CCU',
       'Required age', 'Price', 'DiscountDLC count', 'About the game',
       'Supported languages', 'Full audio languages', 'Reviews',
       'Header image', 'Website', 'Support url', 'Support email', 'Windows',
       'Mac', 'Linux', 'Metacritic score', 'Metacritic url', 'User score',
       'Positive', 'Negative', 'Score rank', 'Achievements', 'Recommendations',
       'Notes', 'Average playtime forever', 'Average playtime two weeks',
       'Median playtime forever', 'Median playtime two weeks', 'Developers',
       'Publishers', 'Categories', 'Genres', 'Tags', 'Screenshots', 'Movies','xxx']
df.head()
```

```python
df['DiscountDLC count'] = None  # 또는 np.nan 으로 바꿔도 됨
df.head()
```

```python
# 이동할 열 리스트 설정
columns_to_shift = df.columns[df.columns.get_loc('DiscountDLC count')+1:]  # DiscountDLC count 다음 열부터 끝까지

# 열 인덱스 기준으로 shift
for i in range(len(columns_to_shift)):
    current_col = columns_to_shift[i]
    prev_col = df.columns[df.columns.get_loc(current_col) - 1]
    df[prev_col] = df[current_col]
df.drop(columns=[columns_to_shift[-1]], inplace=True)
df.head()
```

```python
df_gp = df[df["Average playtime forever"] >= 1][[
    'Name', 'Estimated owners', 'Peak CCU',
    'Metacritic score', 'Positive', 'Negative', 'Average playtime forever']]
df_gp.head()
```

```python
# 범위를 평균값으로 변환하는 함수
def range_to_avg(value):
    try:
        low, high = value.replace(',', '').split(' - ')
        return (int(low) + int(high)) / 2
    except:
        return None  # 혹시 잘못된 값은 결측치 처리

# 적용
df_gp['Estimated owners'] = df_gp['Estimated owners'].apply(range_to_avg)
```

```python
df_gp.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# 1. 사용할 열
features = [
    'Estimated owners', 'Peak CCU', 'Metacritic score',
    'Positive', 'Negative', 'Average playtime forever'
]

# 2. 결측치 제거
df_clean = df_gp.dropna(subset=features).reset_index(drop=True)

# 3. 정규화
scaler = MinMaxScaler()
normalized = scaler.fit_transform(df_clean[features])
df_normalized = pd.DataFrame(normalized, columns=features)

# 4. 가중치 설정
weights = {
    'Estimated owners': 0.30,
    'Peak CCU': 0.1,
    'Metacritic score': 0.05,
    'Positive': 0.15,
    'Negative': -0.15,  # 부정 리뷰는 감점
    'Average playtime forever': 0.30
}

# 5. 가중합 점수 계산
df_normalized['degree of success'] = sum(df_normalized[col] * weight for col, weight in weights.items())

# 6. 점수를 원래 df에 붙이기
df_clean['degree of success'] = df_normalized['degree of success']

# 7. 분위수 기준으로 그룹 나누기
quantiles = df_clean['degree of success'].quantile([0.694, 0.824, 0.924, 0.98]).values

def score_to_group(score):
    if score <= quantiles[0]:
        return 'worst'
    elif score <= quantiles[1]:
        return 'bad'
    elif score <= quantiles[2]:
        return 'common'
    elif score <= quantiles[3]:
        return 'good'
    else:
        return 'goat'

df_clean['success_group'] = df_clean['degree of success'].apply(score_to_group)

# 8. 결과 보기
df_result = df_clean[['Name', 'degree of success', 'success_group'] + features]
df_result.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 그룹별 연령 제한 빈도표 생성
age_dist = df_ch.groupby(['success_group', 'Required age']).size().unstack()

# 각 그룹별 비율로 변환
age_ratio = age_dist.div(age_dist.sum(axis=1), axis=0)  # 행 기준 정규화 (비율)

# 원하는 순서로 인덱스 정렬
group_order = ['goat', 'good', 'common', 'bad', 'worst']
age_ratio = age_ratio.reindex(group_order)

# 스택형 막대그래프 다시 그리기
age_ratio.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Set2')

plt.title('Age required distribution by success group')
plt.xlabel('Success Group')
plt.ylabel('Required age')
plt.legend(title='Required Age', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# 가격 구간 정의
bins = [-0.01, 0, 5, 10, 30, 60, df_ch['Price'].max()]
labels = ['Free', 'Cheap', 'Budget', 'Standard', 'Premium', 'Deluxe']

# 구간에 따라 새로운 컬럼 생성
df_ch['Price Range'] = pd.cut(df_ch['Price'], bins=bins, labels=labels)


# success_group별로 Price Range의 비율 계산
price_dist = df_ch.groupby(['success_group', 'Price Range']).size().unstack().fillna(0)
price_ratio = price_dist.div(price_dist.sum(axis=1), axis=0)

# 원하는 순서로 정렬
group_order = ['goat', 'good', 'common', 'bad', 'worst']
price_ratio = price_ratio.reindex(group_order)

# 시각화
price_ratio.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Set3')
plt.title('Distribution of price by success group')
plt.xlabel('Success Group')
plt.ylabel('Price rate')
plt.legend(title='Price Range', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer
import ast

# 문자열을 리스트로 변환
df_ch['lang_list'] = df_ch['Supported languages'].apply(ast.literal_eval)

# 멀티핫 인코딩
mlb = MultiLabelBinarizer()
lang_dummies = pd.DataFrame(mlb.fit_transform(df_ch['lang_list']),
                            columns=mlb.classes_,
                            index=df_ch.index)

# 언어 열을 원본 데이터에 붙이기
df_lang = pd.concat([df_ch[['Name', 'success_group']], lang_dummies], axis=1)

# success_group별 언어 지원 비율 계산
lang_group_ratio = df_lang.groupby('success_group')[mlb.classes_].mean()

# 그룹별 Top 5 언어 추출
top_languages_per_group = {}

for group in lang_group_ratio.index:
    top5 = lang_group_ratio.loc[group].sort_values(ascending=False).head(5)
    top_languages_per_group[group] = top5

# 시각화용 DataFrame으로 변환
df_top_lang = pd.DataFrame(top_languages_per_group).T

ax = df_top_lang.plot(kind='bar', figsize=(12, 6))

# 제목, 축 라벨 등 추가
plt.title('Top 5 Supported Languages per Success Group')
plt.ylabel('Support Ratio')
plt.xlabel('Success Group')

# 범례 위치 조정
plt.legend(title='Language', bbox_to_anchor=(1.05, 1), loc='upper left')

# 레이아웃 정리
plt.tight_layout()

# 그래프 보여주기
plt.show()
```
